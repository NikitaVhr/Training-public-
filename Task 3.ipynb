## Задание 3
В игре Plants & Gardens каждый месяц проводятся тематические события, ограниченные по времени. В них игроки могут получить уникальные предметы для сада и персонажей, дополнительные монеты или бонусы. Для получения награды требуется пройти ряд уровней за определенное время. С помощью каких метрик можно оценить результаты последнего прошедшего события?\


Предположим, в другом событии мы усложнили механику событий так, что при каждой неудачной попытке выполнения уровня игрок будет откатываться на несколько уровней назад. Изменится ли набор метрик оценки результата? Если да, то как?

## Ответ
В первом случае, для оценки результатов тематического ивента можно воспользоваться следующими метриками:


_Если нас интересует финансовый результат_


* **Revenue** - выручка. Поможет увидеть сколько мы вообще заработали за ивент. Результаты можно сравнить с другими месячными событиями, чтобы выявить наиболее интересные для игроков темы.


* **ARPDAU** - доход, который в среднем приносит один пользователь приложения за один день. Поможет узнать стали ли мы больше зарабатывать во время тематического события с одного активного игрока. Можно также отдельно посмотреть ARPU и DAU, чтобы понять за счет конкретно чего увеличился / уменьшился доход.


* **Paying Conversion** - % игроков, совершивших покупку, из тех, кто установил приложение за определенный период.
Поможет посмотреть влияние ивента на конверсию новых игроков. В ходе анализа новых пользователей можно разделить на когорты по дням.


_Если нас интересует статистика связанная с дизайном и реализацией ивента_


* **Конверсия N-го уровня** - % игроков, дошедших до N-го уровня от всех принявших участие в ивенте.
Поможет понять заинтересованность в ивенте, увидеть отвалы игроков и поможет отрегулировать сложность уровней для будущих ивентов.


* **Total Daily Play Time**  - суммарная продолжительность всех сессий в день поделенная на количество активных пользователей в этот день. Поможет узнать сколько в среднем времени в день проводят игроки  в приложении. По идее, если уровни идут в нагрузку к основному контенту, то метрика должна увеличиться.


* **Среднее время прохождения уровня** - говорящая сама за себя метрика. Поможет выявить наиболее сложные уровни на которых можно будет добавить предложение о покупке валюты в аналогичных будущих ивентах.


* **Completes** - показывает сколько раз игроки завершили тот или иной уровень игры. Поможет отследить прогрессию игроков как после, так и во время ивента и если нам, например, покажется, что она слишком быстрая или медленная, мы можем скорректировать сложность еще не пройденных уровней.


* **Concurrent Users** - среднее число одновременно играющих пользователей. Поможет оценить нагрузку на сервера, если, например, во время ивента были жалобы на соединение от игроков.


* **Error count** - число ошибок за определенное время. Поможет понять насколько хорошо был реализован ивент технически.

Во втором случае (усложнили механику событий так, что при каждой неудачной попытке выполнения уровня игрок будет откатываться на несколько уровней назад), к перечисленным метрикам выше я бы еще добавил:


* **Fails** - показывает сколько раз игроки провалили любой уровень в день. Данная метрика поможет выявить наиболее сложные уровни, на которых можно будет добавить предложение с покупкой внутриигровой валюты чтобы, например, пропустить уровень. Еще это метрика поможет выявить чрезмерно сложные уровни, которые слишком раздражают игроков и требуют упрощения.


* **Win percentage** - показывает долю успешных попыток от всех попыток пройти уровень. Как и fails поможет выявить наиболее сложные для игроков уровни.
